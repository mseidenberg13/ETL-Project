```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extract the kaggle_income.csv file, converted the zip code to a string to keep leading zeros

income_file = open(r"C:\Users\mseid\UCBWork\ETL-Project\Resources\kaggle_income.csv")
income_df = pd.read_csv(income_file, converters={'Zip_Code': lambda x: str(x)})
income_df
```

```python
#Transform the income table by dropping and renaming columns

new_income_df = income_df[["State_ab", "County", "City", "Zip_Code", "Area_Code", "Mean", "Median", "Stdev"]]
new_income_df = new_income_df.rename(columns={"State_ab": "State", "Stdev": "Standard_Deviation_Income", "Mean": "Mean_Income", "Median": "Median_Income"})

new_income_df.drop_duplicates("Zip_Code", inplace=True)

new_income_df
```

```python
# Set index to Zip_Code

new_income_df = new_income_df.set_index("Zip_Code")
new_income_df
```

```python
#Extract the housing tabel from the Zillow csv file, converted the zip code to a string to keep leading zeros

housing_file = (r"C:\Users\mseid\UCBWork\ETL-Project\Resources\Zip_Zhvi_AllHomes.csv")
housing_df = pd.read_csv(housing_file, converters={'RegionName': lambda x: str(x)})
housing_df.head()
```

```python
#Transform the income table by dropping and renaming columns

new_housing_df = housing_df[["RegionName", "State", "City", "CountyName", "2020-03-31"]]
new_housing_df = new_housing_df.rename(columns={"RegionName": "Zip_Code", "CountyName": "County", "2020-03-31": "2020_average_house_price"})

new_housing_df
```

```python
#Set index to Zip_Code

new_housing_df = new_housing_df.set_index("Zip_Code")
new_housing_df
```

```python
#Merge the income and housing tables together on Zip Code to create a new table that shows income and housing prices together
#where there is data that matches from both tables

income_housing_df = new_housing_df.merge(new_income_df, how = 'inner', left_index=True, right_index=True)

income_housing_df
```

```python
#Transform the combined table by dropping and renaming columns

new_income_housing_df = income_housing_df[["State_x", "City_x", "County_x", "2020_average_house_price", "Area_Code", "Mean_Income", "Median_Income", "Standard_Deviation_Income" ]]
new_income_housing_df = new_income_housing_df.rename(columns={"State_x": "State", "City_x": "City", "County_x": "County"})
new_income_housing_df
```

```python
#Extract the restuarant data into a pandas data frame, converted the zip code to a string to keep leading zeros

eats_file = open(r"C:\Users\mseid\UCBWork\ETL-Project\Resources\yellowpages_com-restaurant_sample.csv")
eats_df = pd.read_csv(eats_file, converters={'Zip Code': lambda x: str(x)})
eats_df.head()
```

```python
#Transform the eats table by dropping and renaming columns, set index to Zip_Code

new_eats_df = eats_df[["Name", "Street", "Zip Code", "City", "State", "Phone", "Categories"]]
new_eats_df = new_eats_df.rename(columns={"Name": "Restaurant_Name", "Street": "Restaurant_Address", "Zip Code": "Zip_Code", "Categories": "Restaurant_Categories"})
new_eats_df = new_eats_df.set_index("Zip_Code")

new_eats_df
```

```python
#Join the income_housing_df with the eats_df to create a new table with income, housing, and retaurant data

income_housing_eats_df = new_income_housing_df.merge(new_eats_df, how = 'inner', left_index=True, right_index=True)

income_housing_eats_df
```

```python
#Transform the joined table by dropping and renaming columns

new_income_housing_eats_df = income_housing_eats_df[["State_x", "City_x", "County", "2020_average_house_price", "Area_Code", "Mean_Income", "Median_Income", "Standard_Deviation_Income", "Restaurant_Name", "Restaurant_Address", "Phone", "Restaurant_Categories"]]
new_income_housing_eats_df = new_income_housing_eats_df.rename(columns={"State_x": "State", "City_x": "City"})

new_income_housing_eats_df.head()
```

```python
#Open population file and convert zip code to string to keep the 0's

pop_file = open(r"C:\Users\mseid\UCBWork\ETL-Project\Resources\population_by_zip_2010.csv")
pop_df = pd.read_csv(pop_file, converters={'zipcode': lambda x: str(x)})
pop_df.head()
```

```python
#Clean population by extracting certail columns, renaming columns, dropping duplicate zip codes, and setting index to zip code

new_pop_df = pop_df[["Ã¯Â»Â¿population", "zipcode"]]
new_pop_df = new_pop_df.rename(columns={"Ã¯Â»Â¿population": "Population", "zipcode": "Zip_Code"})
new_pop_df.drop_duplicates("Zip_Code", inplace=True)
new_pop_df = new_pop_df.set_index("Zip_Code")

new_pop_df
```

```python
#Join the income_housing_eats_df with population to create a joined table

income_housing_eats_pop_df = new_income_housing_eats_df.merge(new_pop_df, how = 'inner', left_index=True, right_index=True)

income_housing_eats_pop_df
```

```python
#Join the income_housing_df with population to create a joined table without restaurants

income_housing_pop_df = new_income_housing_df.merge(new_pop_df, how = 'inner', left_index=True, right_index=True)

income_housing_pop_df
```

```python
#Create a connection to postgres

rds_connection_string = "postgres:postgres@localhost:5432/income_housing_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```
